```python
import pandas as pd
import numpy as np
```

```python
df_all_series = pd.read_excel('base_all_series.xlsx')
df_all_series
```

```python
#países en desarrollo
df_desarrollo = pd.read_excel('paises_via_desarrollo.xls', sheet_name='Metadata - Countries')
df_desarrollo
```

```python
countries_not_dev = df_desarrollo[df_desarrollo['IncomeGroup'] != 'High income']
countries_not_dev
```

```python
countries = countries_not_dev['Country Code'].unique().tolist()
print(countries)
len(countries)
```

```python
#filtrar el dataframe solo por los países en vía de desarrollo

df_filtrado = df_all_series[df_all_series['Country Code'].isin(countries)]
df_filtrado
```

```python
# Transformar las columnas de años en filas

df_years = df_filtrado.drop(['Series Code'], axis=1)

df_years = df_years.drop(df_years[df_years['Series Name'].isna()].index)
df_years = df_years.drop(df_years[df_years['Country Name'].isna()].index)


df_years = df_years.melt(
    id_vars=['Country Name', 'Country Code', 'Series Name'],
    var_name='Year',
    value_name='Value'
)
df_years
```

```python
# Limpiar la columna "Year" para extraer solo el número del año
df_years['Year'] = df_years['Year'].str.extract('(\d{4})').astype(int)

df_years
```

```python
#Convertir a número
df_years['Value'] = pd.to_numeric(df_years['Value'], errors='coerce')
df_years
```

```python
#tabla pivote
df_pivot = df_years.pivot_table(
    index=['Country Name', 'Country Code', 'Year'],  # Identificadores únicos
    columns='Series Name',                          # Categorías convertidas en columnas
    values='Value',                                 # Valores de las columnas
    aggfunc='first'                                 # Manejar duplicados (si los hay)
).reset_index()
df_pivot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un DataFrame booleano donde True significa que el valor es nulo
plt.figure(figsize=(20, 6))
sns.heatmap(df_pivot.isnull(), cbar=False, cmap='viridis', xticklabels=True)

plt.title('Mapa de Calor de Valores Nulos')
plt.show()
```

```python
#eliminar filas con valor nulo en el PIB

df = df_pivot.dropna(subset=['GDP per capita (current US$)'])
df
```

```python
plt.figure(figsize=(20, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', xticklabels=True)

plt.title('Mapa de Calor de Valores Nulos')
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

# Variables predictoras y objetivo
variables = [
    'Population growth (annual %)',
    'Access to electricity (% of population)',
    'Foreign direct investment, net inflows (% of GDP)',
    'Government expenditure on education, total (% of GDP)'
]
target = 'GDP per capita (current US$)'

# Configurar el número de filas y columnas
num_variables = len(variables)
num_columns = 2  # Dos gráficos por fila
num_rows = math.ceil(num_variables / num_columns)  # Calcular filas necesarias

# Crear subplots
fig, axes = plt.subplots(num_rows, num_columns, figsize=(12, num_rows * 5))

# Aplanar los ejes si hay más de una fila
axes = axes.flatten()

# Iterar sobre cada variable y crear gráficos de dispersión
for i, variable in enumerate(variables):
    axes[i].scatter(df[variable], df[target], alpha=0.7, color='blue')
    axes[i].set_title(f'{variable} vs {target}', fontsize=12)
    axes[i].set_xlabel(variable, fontsize=10)
    axes[i].set_ylabel(target, fontsize=10)
    axes[i].grid(True)

# Desactivar los ejes no usados (si hay un número impar de variables)
for j in range(len(variables), len(axes)):
    axes[j].axis('off')

# Ajustar el diseño del lienzo
plt.tight_layout()
plt.show()
```

```python
# Configurar el número de filas y columnas
num_variables = len(variables)
num_columns = 2  # Dos gráficos por fila
num_rows = math.ceil(num_variables / num_columns)

# Crear subplots para histogramas
fig, axes = plt.subplots(num_rows, num_columns, figsize=(12, num_rows * 5))

# Aplanar los ejes si hay más de una fila
axes = axes.flatten()

# Generar histogramas
for i, variable in enumerate(variables):
    axes[i].hist(df[variable].dropna(), bins=30, edgecolor='black', alpha=0.7)
    axes[i].set_title(f'Histograma de {variable}', fontsize=12)
    axes[i].set_xlabel(variable, fontsize=10)
    axes[i].set_ylabel('Frecuencia', fontsize=10)

# Desactivar los ejes no usados
for j in range(len(variables), len(axes)):
    axes[j].axis('off')

# Ajustar el diseño del lienzo
plt.tight_layout()
plt.show()
```

```python
# Configurar el número de filas y columnas
num_variables = len(variables)
num_columns = 2  # Dos gráficos por fila
num_rows = math.ceil(num_variables / num_columns)

# Crear subplots para boxplots
fig, axes = plt.subplots(num_rows, num_columns, figsize=(12, num_rows * 5))

# Aplanar los ejes si hay más de una fila
axes = axes.flatten()

# Generar boxplots
for i, variable in enumerate(variables):
    sns.boxplot(data=df, y=variable, ax=axes[i], color='skyblue')
    axes[i].set_title(f'Boxplot de {variable}', fontsize=12)
    axes[i].set_xlabel('')
    axes[i].set_ylabel(variable, fontsize=10)

# Desactivar los ejes no usados
for j in range(len(variables), len(axes)):
    axes[j].axis('off')

# Ajustar el diseño del lienzo
plt.tight_layout()
plt.show()
```
